## 1. Executive Summary

During 2017-05-04 and 2017-05-22, an AB test was conducted to test a new experience on the Super Math Saga game, to find out the impact of the said experience, with 80% of the users tested assigned to control group A, and 20% of the users tested assigned to treatment group B. The treatment group received the new experience while the control group did not. 

While performing sanity checks for the analysis, we have confirmed to use playerid as the unit of randomization. However, looking at the activity table, we have realized there is **not sufficient data for all players**. Specifically, **only data on players with activity during the test period has been recorded, for the players that have installed the game but has not performed any activity (played games, purchased hints), there was no record. This could cause potential bias in our analysis**. However, we carried on the experiment with the data at hand.

Through the analysis we were able to find out: 
1. The treatment **did not show any statistically significant change in conversion rate**.
2. The treatment **did not show any statistically significant change in retention rate**.
3. The treatment **has a negative effect on engagement by decreasing the average game rounds played per day per player**, control group seems to have higher engagement.
4. The treatment has **a positive effect on revenue by increasing the average revenue per daily active user, but this seems to be due to a novely effect, which fades over time**. 
5. When segmenting the players into new and old players, **both showed less engagement towards the treatment. While, old players are likely to spend more than new players under the treatment**.

We have also suggested the new experience implemented could be:

**An increased difficulty of the math puzzles to solve to pass the levels**.\
Because of an increased difficulty, players will be less likely to pass levels and more likely to generate frustration towards the game, resulting in less engagement by playing less game rounds. \
For the old players, since they are experiencing more difficult game rounds, they will tend to lean towards passing levels as before, before realizing the difficulty has increased, they will be incentivised to purchase hints to pass the levels. However, this is only a novelty effect which will fade as soon as they realize the game design has become more difficult.


In conclusion,after implementing the new experience, we have observed a negative impact on engagement, no significant changes in conversion, retention, and uplift in revenue but only due to novelty effect, which results in an **overall negative impact on user experience**.Through the research we were able to observe when releasing a new experience, old users tend to spend more in the game and result in a positive revenue uplift. **This implies the possibility for us to conduct further tests on spending, and periodically testing impacts of new experiences. If we can avoid negatively impacting engagement and retention, we might be able to increase overall revenue with releasing new experiences**.

**However, we cannot be sure if this is the real case during this experiment since players with no activity during the test period but have been assigned to groups were not recorded**.This could cause bias to our engagement (average game rounds per player has played per day) analysis since we do not record the activity of all players. **We would suggest to collect further data, make sure all activity status and inactivity status of all playerids assigned are recorded and conduct the analysis again or extend the test period while recording all data**.

## 4. Analysis

For each evaluation metric, the difference between the values from both groups is measured. Then, we compute the p-value to see if it's smaller than the critical value given significance level at 0.05.

### 4.1 Conversion rate

Conversion rate implies the rate of which players convert to paying members and contribute towards our revenue. Here we noticed some users have converted before the assignment date, we exclude these purchases out of the conversion, and only compare conversions that have happened after the assignments.\
As conversion is a categorical variable (which means players are either converted or not converted), we would like to find out if having the treatment actually has any association with the results of conversion, therefore we would use a Fisher Exact test. If the p value is smaller than 0.05, we can conclude having the treatment has significant changes on conversion rate.\
Our Hypothesis are as followed:\
**H0**: Conversion rates and having the treatment or not are independent.\
**H1**ï¼Conversion rates and having the treatment or not are not independent.

Of the 8265610 entrants of group A, around 0.397% players have converted, of the 2065446 entrants of group B, around 0.405% players have converted.

The p-value abtained is around 0.099, higher than the significance level 0.05, we fail to reject the null hypothesis.\
The odds ratio indicate that the proportions of the control group is roughly 0.98 times that of the the treatment group, also supporting the fact **the conversion rate between two groups do not seem to have a significant difference**.

## 4.2 Retention Rate 

Retention rate refers to the numbers of days players have been active throughout the period.\
This metric evaluates how long the experiences of the game can hold players' attention, ideally we would prefer the retention rate to be bigger. According to central limit theorem, when the sample size is large enough, we assume the mean of retention rate to follow a normal distribution. Therefore we use a z test.\
Our hypothesis are as followed:\
**H0**: Retention rate in treatment group has no significant difference from control group.\
**H1**: Retention rate in treatment group has a significant difference from control group.

p-value is larger than the 0.05 threshold, we fail to reject the null hypothesis. **No significant changes can be detected in retention rate of the total period between two groups**.

## 4.3 Engagement
Engagement evaluates how good we are in keep our players activated and playing the games, this is an important metric since staying active is the first step to conversion. We evaluate engagment in a twofold method.
### 4.3.1 Average game rounds player per player
We look at the average games each player in two groups have played in the testing period. Similarly we assume normal distribution for average game rounds according to central limit theorem and perform a z test.\
Our Hypothesis are as followed:\
**H0**: Average game ends per player per day in treatment group has no significant difference from control group.\
**H1**: Average game ends per player per day in treatment group has a significant difference from control group.

Based on the calculated results,we reject the null hypothesis, and the average gameends per player per day has shown significant difference. We also observe that **the treatment group plays 0.2336 less game ends than control group on average, indicating a negative impact from the new experience**.

We can see that after the experiment starts on 2017-05-04 , **treatment has seen dropping in average game rounds played per player since the experiment starts, while the control group has stayed stable**, which is an indication of the treatment has lowered the engagement.

## 4.4 Revenue
### 4.4.1 Average revenue per day per active user
We look at the average revenue per day per active user, which is measured by the average number of purchases per day per active user in two groups. Similarly we assume normal distribution for average number of purchases per day per active user according to central limit theorem and perform a z test.\
Our Hypothesis are as followed:\
**H0**: average number of purchases per day per active user in treatment group has no significant difference from control group.\
**H1**: average number of purchases per day per active user in treatment group has a significant difference from control group.

Based on the calculated results,we reject the null hypothesis, and the average purchase per day per active player has shown significant difference. We also observe that **the treatment group places 0.0019 more purchases than control group on average, indicating a positive impact on revenue from the new experience**.

We can see **an initial increase in Average Revenue per Daily Active user in the treatment group, the uplift seems to fade as the time period extends**. \
This phenomenon inspired us there could be a possibility of **novelty effect**, where old users see a new experience and since it has never been seen before, they are triggered to place more purchases, we will analyse this in the next part.

## 5. Breakdown Analysis of players

In light of findings in 4.4.1, we now look at if there is novelty effect existing in the experiment. Therefore, we segment the players into old players and new players. We define new players as players who installed the game on the same day they were assigned to a group, and old players as players who installed the game before they were assigned to a group.

Based on the calculated results,we fail to reject the null hypothesis, and **the average purchase per daily active player among new players has shown no significant difference**.

Based on the calculated results,we reject the null hypothesis, and the average purchase per daily active player in old players has shown significant difference. We also observe that **the old players in treatment group places 0.00227 more purchases than the old players in control group on average, indicating a positive impact on revenue from the new experience**.

Based on the calculated results, we reject both the null hypothesis, and both groups have shown significant difference in average game rounds played per day, and the impact is negative. This is **consistent with our findings in part 4 and no novelty effect was detected in average game rounds played per player**.

Based on above graphs we can also see old players with treatment has seen a consistent uplift in revenue then control, while new players revenue seem to fluctuate.\
The new players' increase in Average Revenue per Daily Active User, when exposed to the treatment is not significant. **While the old players spending is significant. we conclude that the old players are spending due to a novelty effect**.
